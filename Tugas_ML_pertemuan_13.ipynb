```python
import pandas as pd

from google.colab import files
uploaded = files.upload()   # upload file

df = pd.read_csv('/content/data.csv')  # sesuaikan dengan nama file
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1, errors='ignore')
df.head()
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y[:10]
```

```python
from sklearn.preprocessing import StandardScaler

# Membuat objek scaler
scaler = StandardScaler()

# Melakukan scaling pada data fitur (X)
X_scaled = scaler.fit_transform(X)

# Menampilkan bentuk dataset hasil scaling
print(X_scaled.shape)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("Ukuran sebelum PCA:", X_scaled.shape)
print("Ukuran sesudah PCA:", X_pca.shape)
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Explained:", pca.explained_variance_ratio_.sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Breast Cancer Dataset')
plt.colorbar()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print("Akurasi Model :", accuracy)
```

```python

from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:\n", cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix PCA + Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
